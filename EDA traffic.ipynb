```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
traffic_stops = pd.read_csv('/Jupyter/Temp/ar_little_rock_2020_04_01.csv')
traffic_stops.head(10)
```

```python
fig, ax = plt.subplots() # this allow us to create multiple plots at once however here we are only creating one. to create more we can use 'plt.subplots(rows, columns)'
ax.hist(traffic_stops['subject_age'], linewidth = 10, edgecolor = 'white') # this will create a histogram for column named 'subject_age' in 'traffic_stops'
ax.set(xlim = (0, 100), xticks = np.linspace(0, 90, 11), ylim = (0, 4000), yticks = np.linspace(0, 3500, 9)) # here we are customizing the x & y axis of our graph using 'xlim' & 'ylim' to limit the data that is going to get plotted i.e. if there is an row in data for which the age is more than '100' it will be ignored then we use 'xticks' & 'yticks' to give the ticks some desired values i.e. 0, 9, 18, so on for x-axis and 0 to 3500 on y-axis
plt.show() # this will plot the graphs stated above in order they are stated
```

```python
plt.hist(traffic_stops['subject_age'], color='skyblue', edgecolor='black') # here unlike above we are simply creating one single histogram plot with no scope to plot another graph alongside it
plt.title('Histogram of subject_age') # this allow us to change the tittle of our graph that will be plotted
plt.xlabel('subject_age') # this to change the label on x-axis
plt.ylabel('Frequency') # and this to change the label on y
plt.show() # this will plot the graph
```
